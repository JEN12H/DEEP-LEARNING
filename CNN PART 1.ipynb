```python
import tensorflow as tf
from tensorflow import keras
from keras.layers import Dense,Flatten,Conv2D,MaxPooling2D,AveragePooling2D
from keras import Sequential
from keras.datasets import mnist
```

```python
(X_train,y_train),(X_test,y_test) = mnist.load_data()
```

```python
model = Sequential()
model.add(Conv2D(32,kernel_size=(3,3),padding='valid',activation='relu',input_shape=(28,28,1)))# PADDING VALID MEANS WE ARE NOT USING IT . IF WE DO PADDING = SAME THEN PADDING IS AUTOMATICALLY APPLIED
model.add(Conv2D(32,kernel_size=(3,3),padding='valid',activation='relu')) # 32 is NUMBER OF FILTERS THAT WE ARE USING HERE KERNEL SHAPE IS THE SHAPE OF FILTER
model.add(Conv2D(32,kernel_size=(3,3),padding='valid',activation='relu'))
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model = Sequential()
model.add(Conv2D(32,kernel_size=(3,3),padding='same',activation='relu',input_shape=(28,28,1)))# PADDING VALID MEANS WE ARE NOT USING IT . IF WE DO PADDING = SAME THEN PADDING IS AUTOMATICALLY APPLIED
model.add(Conv2D(32,kernel_size=(3,3),padding='same',activation='relu')) # 32 is NUMBER OF FILTERS THAT WE ARE USING HERE KERNEL SHAPE IS THE SHAPE OF FILTER
model.add(Conv2D(32,kernel_size=(3,3),padding='same',activation='relu'))
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
# IN CASE OF PADDING = SAME NO INFORMATION IS LOST FROM THE IMAGE.
```

```python
model = Sequential()
model.add(Conv2D(32,kernel_size=(3,3),padding='same',strides=(2,2),activation='relu',input_shape=(28,28,1)))# PADDING same MEANS WE ARE NOT USING IT . IF WE DO PADDING = SAME THEN PADDING IS AUTOMATICALLY APPLIED
model.add(Conv2D(32,kernel_size=(3,3),padding='same',strides=(2,2),activation='relu')) # 32 is NUMBER OF FILTERS THAT WE ARE USING HERE KERNEL SHAPE IS THE SHAPE OF FILTER
model.add(Conv2D(32,kernel_size=(3,3),padding='same',strides=(2,2),activation='relu'))
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model = Sequential()
model.add(Conv2D(32,kernel_size=(3,3),padding='valid',strides=(2,2),activation='relu',input_shape=(28,28,1)))# PADDING valid MEANS WE ARE NOT USING IT . IF WE DO PADDING = valid THEN PADDING IS AUTOMATICALLY APPLIED
model.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))
model.add(Conv2D(32,kernel_size=(3,3),padding='valid',strides=(2,2),activation='relu')) # 32 is NUMBER OF FILTERS THAT WE ARE USING HERE KERNEL SHAPE IS THE SHAPE OF FILTER
model.add(MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'))

model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

# LeNet ARCHITECTURE

```python
model = Sequential()
model.add(Conv2D(6,kernel_size=(5,5),padding='valid',activation='tanh',input_shape=(32,32,1)))
model.add(AveragePooling2D(pool_size=(2,2),strides=2,padding='valid'))
model.add(Conv2D(16,kernel_size=(5,5),padding='valid',activation='tanh'))
model.add(AveragePooling2D(pool_size=(2,2),strides=2,padding='valid'))
model.add(Flatten())
model.add(Dense(120,activation='tanh'))
model.add(Dense(84,activation='tanh'))
model.add(Dense(10,activation='softmax'))
```
