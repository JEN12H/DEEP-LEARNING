```python
import zipfile
zip_ref = zipfile.ZipFile('/content/dogs-vs-cats.zip','r')
zip_ref.extractall('/content')
zip_ref.close()
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,Flatten
from keras.applications.vgg16 import VGG16
```

```python
conv_base = VGG16(
    weights = 'imagenet',
    include_top = False,
    input_shape = (150,150,3)
)
```

```python
conv_base.summary()
```

```python
# HERE WE HAVE NOT INCLUDED DENSE LAYER IN OUR MODEL .
# FLATTEN LAYERS ACTS AS RESHAPE THAT TURNS SPATIAL FEATURE MAPS INTO A FORMAT THAT A DENSE LAYER CAN UNDERSTAND.
```

```python
model = Sequential()
model.add(conv_base) # WE USED PREVIOUS MODEL WEIGHTS THAT IS BASE ON   IMAGE NET DATA.
model.add(Flatten())
model.add(Dense(256,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
conv_base.trainable = False
# WE DID IT FALSE AS WE DONT WANT TO AGAIN TRAIN THE CONVOLUTION LAYERS .WE HAVE FREEZED THE CONV LAYERS OF OUR MODEL.
```

```python
train_ds = keras.utils.image_dataset_from_directory(
    directory = '/content/train',
    labels = 'inferred',
    label_mode = 'int',
    batch_size=32,
    image_size = (150,150)
)
validation_ds= keras.utils.image_dataset_from_directory(
    directory = '/content/test',
    labels = 'inferred',
    label_mode = 'int',
    batch_size = 32,
    image_size = (150,150)
)
```

```python
# NORMALIZATION OF PIXEL VALUES TO BRING IT IN RANGE OF 0 and 1
```

```python
def process(image,label):
  image = tf.cast(image/255.,tf.float32)
  return image,label
train_ds = train_ds.map(process)
valdiation_ds = validation_ds.map(process)
```

```python
model.compile(optimizer = 'adam',loss = 'binary_crossentropy',metrics = ['accuracy'])
```

```python
history = model.fit(train_ds, epochs = 10,validation_data = validation_ds)
```
