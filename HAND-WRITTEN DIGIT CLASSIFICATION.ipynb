```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
(X_train,y_train),(X_test,y_test) = keras.datasets.mnist.load_data()
```

```python
X_train.shape
```

```python
X_train[0]
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_train[1])
```

```python
X_train = X_train/255
X_test = X_test /255
```

```python
X_train[1]
```

```python
model = Sequential()
model.add(Flatten(input_shape=(28,28))) # FLATTEN IS USED TO BRING HIGH DIMENSIONAL ARRAYS INTO 1-D REPRESENTATION
model.add(Dense(128,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(10,activation='softmax')) # SOFTMAX IS USED FOR MULTI-CLASS CLASSIFICATION
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam') # WE DONT NEED TO ONE-HOT ENCODE OUR LABELS IN SPARSE_CATEGORICAL CROSSENTROPY UNLIKE CATEGORICAL ENTROPY
```

```python
history = model.fit(X_train,y_train,epochs = 25,validation_split = 0.2)
```

```python
y_prob = model.predict(X_test)
```

```python
y_pred = y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```
