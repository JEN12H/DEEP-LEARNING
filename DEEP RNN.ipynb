```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,SimpleRNN,Embedding,LSTM,GRU
```

```python
(X_train,y_train), (X_test,y_test) = imdb.load_data(num_words=10000)
X_train = pad_sequences(X_train,maxlen=50)
X_test = pad_sequences(X_test,maxlen=50)
```

```python
X_train[0]
```

```python
model = Sequential([
    Embedding(10000,32), # EMBEDDING LAYER TO CONVERT WORDS TO VECTORS .EVERY WORD WILL BE REPRESENTED AS A 32 DIMENSIONAL NUMBER
    SimpleRNN(5,return_sequences=True), # RNN LAYERS WITH 5 UNITS
    SimpleRNN(5), # ANOTHER RNN LAYER WITH 5 UNITS
    Dense(1,activation='sigmoid') # OUTPUT LAYER FOR BINARY CLASSFICATION
])
model.build((None,100))
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(X_train,y_train,epochs=10,validation_data=(X_test,y_test))
```

```python
# SIMILARLY DEEP LSTM
model = Sequential([
    Embedding(10000,32),
    LSTM(5,return_sequences=True),
    LSTM(5),
    Dense(1,activation='sigmoid')
])
model.build((None,100))
model.summary()
```

```python
# SIMILARLY DEEP GRU
model = Sequential([
    Embedding(10000,32),
    GRU(5,return_sequences=True),
    GRU(5),
    Dense(1,activation='sigmoid')
])
model.build((None,100))
model.summary()
```
